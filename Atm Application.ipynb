```python
users = {"pin": [1234, 1235, 1236, 1237, 1238, 1239], "account_num": [10001, 10002, 10003,10004, 10005, 10006], 
         "balance": [10000, 15000, 20000, 25000, 30000, 40000]}

actions = {"1": "View Balance", "2":"Withdraw Balance", "3":"Deposit Balance", "4":"Change Pin", 
           "5":"Transfer Balance", "6":"Exit"}
def display_menu():
    print("\n*****Welcome to Atm*****\n")
    for action in actions:
        print(f"({action}) {actions[action]}")
def view_balance(pin):
    if pin in users["pin"]:
        index = users["pin"].index(pin)
        print(f"Your Balance: {users['balance'][index]}")
    else:
        print("Please Enter valid Pin")
def withdraw_balance(pin, amount):
    if pin in users["pin"]:
        index = users["pin"].index(pin)
        if users['balance'][index] >= amount:
            users['balance'][index] -= amount
            print(f"{amount} successfully withdrawn from your account")
        else:
            print("You have insufficient balance")
    else:
        print("Please Enter valid pin")
def deposit_balance(pin, amount):
    if pin in users["pin"]:
        index = users["pin"].index(pin)
        if amount > 0:
            users['balance'][index] += amount
            print(f"{amount} succesffuly added into your account")
        else:
            print("Please Enter valid amount for deposit")
    else:
        print("Your pin is invalid, please try again")
def transfer_balance(pin, receiver_account_num, amount):
    if pin in users["pin"]:
        sender_index = users["pin"].index(pin)
        if amount <= users["balance"][sender_index]:
            if receiver_account_num in users["account_num"]:
                receiver_index = users["account_num"].index(receiver_account_num)
                users["balance"][sender_index] -= amount
                users["balance"][receiver_index] += amount
                print(f"{amount} successfully added into the receiver account")
            else:
                print("Invalid receiver account number, please try again")
        else:
            print("Your have insufficient balance")
    else:
        print("Invalid Pin, Please try again")
def change_pin(pin):
    if pin in users["pin"]:
        new_pin = int(input("Please enter new pin: "))
        index = users["pin"].index(pin)
        users['pin'][index] = new_pin
        print("Pin successfully changed")
    else:
        print("Please enter valid pin")
def user_details(pin_prompt, amount_prompt = None, receiver_account_prompt = None):
    
    if not amount_prompt and not receiver_account_prompt:
        pin = int(input(pin_prompt))
        return pin
    elif amount_prompt and not receiver_account_prompt:
        pin = int(input(pin_prompt))
        amount = int(input(amount_prompt))
        return pin, amount
    else:
        pin = int(input(pin_prompt))
        amount = int(input(amount_prompt))
        receiver_account_num = int(input(receiver_account_prompt))
        return pin, amount, receiver_account_num
def atm():
    pin_prompt = "Enter your pin:"
    amount_prompt = "Enter your amount: "
    receiver_account_prompt = "Enter reciever account number"
    flag = True
    while flag:
        display_menu()
        user_choice = int(input("Enter your choice: "))
        if user_choice == 1:
            pin = user_details(pin_prompt)
            view_balance(pin)
        elif user_choice == 2:
            pin,amount = user_details(pin_prompt, amount_prompt)
            withdraw_balance(pin, amount)
        elif user_choice == 3:
            pin,amount = user_details(pin_prompt, amount_prompt)
            deposit_balance(pin, amount)
        elif user_choice == 4:
            pin = user_details(pin_prompt)
            change_pin(pin)
        elif user_choice == 5:
            pin, amount, receiver_account = user_details(pin_prompt, amount_prompt,  receiver_account_prompt,)
            print("Receiver Account:", receiver_account)
            transfer_balance(pin, receiver_account, amount)
        elif user_choice == 6:
            flag = False
            
            
atm()
```
